```python
import seaborn as sns
from scipy import stats
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

```python
formula = "Y ~ CR + DA + FL + OPM + RT + TAT + DE + PT"
mod = smf.ols(formula=formula, data=data)
res = mod.fit()
res.summary()
```

novy model bez statisticky nevyznamnych koeficientu

```python
formula2 = "Y ~ FL + OPM + TAT"
mod = smf.ols(formula=formula2, data=data)
res = mod.fit()
res.summary()
```

```python
data["Y_predicted"] = res.predict(data)
data.head()
```

```python
data["Y_difference"] = data["Y_predicted"] - data["Y"]
sorted_data = data.sort_values(by="Y_difference", ascending=False)
sorted_data.head(1)
```

Bonus 1

```python
formula3 = "Y ~ CR + DA + FL + OPM + RT + TAT + DE + PT"
mod = smf.rlm(formula=formula3, data=data)
res = mod.fit()
res.summary()
```

```python
formula4 = "Y ~ CR + FL + OPM + TAT"
mod = smf.rlm(formula=formula4, data=data)
res = mod.fit()
res.summary()
```

```python
data["Y_predicted_robust"] = res.predict(data)
data["Y_difference_robust"] = data["Y_predicted_robust"] - data["Y"]
sorted_data2 = data.sort_values(by="Y_difference_robust", ascending=False)
sorted_data2.head(1)
```

Ano, jedna se o stejnou akcii.

Bonus 2

```python
from statsmodels.stats.outliers_influence import OLSInfluence #Cook distance and other parameters, cookova je [0]

formula2 = "Y ~ FL + OPM + TAT"
mod = smf.ols(formula=formula2, data=data)
res = mod.fit()
influence = OLSInfluence(res)
data["Cook_distance"] = influence.cooks_distance[0]
data.head()
```

```python
data_cook = data[data["Cook_distance"] <= 1]
data_cook.head()
```

```python
formula2 = "Y ~ FL + OPM + TAT"
mod = smf.ols(formula=formula2, data=data_cook)
res = mod.fit()
res.summary()
```

```python
data_cook["Y_predicted_cook"] = res.predict(data_cook)
data_cook["Y_difference_cook"] = data_cook["Y_predicted_cook"] - data_cook["Y"]
sorted_data3 = data_cook.sort_values(by="Y_difference_cook", ascending=False)
sorted_data3.head(1)
```

Vyslo mi, ze jde o stejnou akcii jako u predchozich modelu.
